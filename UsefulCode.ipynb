## 1. Cosine similarity(코사인 유사도) 계산

```python
import numpy as np

def cosine_similarity(a,b):
  return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

## 1. 그래프 그리기

```python
import matplotlib.pyplot as plt

search_list = ['HYUNDAI','LG','POSCO','SAMSUNG','SHINHAN', 'SK']
color_list =['blue','green','cyan','gray','teal','lime']

result_dict = {}
for nm in search_list:
  filtered_list = raw[raw['sci_grp_nm'].str.contains(nm)]
  result_dict[nm]=filtered_list

plt.rc('font', family='Malgun Gothic')
plt.title('LE_TOP6_GROUP_SACCR_EAD')

for nm, colors in zip(search_list,color_list):
  plt.plot(result_dict[nm]['date'],result_dict[nm]['saccr_ead'],label=nm, color=colors)

plt.xlabel('date')
plt.ylabel('EAD')
plt.legend(loc='upper left')
plt.show()
```
